# Seasonal monthly error metrics

```python
pretrain_monthly_error1 = pd.read_csv("../river-dl/pretrain_1/seasonal_month_metrics.csv")
pretrain_monthly_error2 = pd.read_csv("../river-dl/pretrain_2/seasonal_month_metrics.csv")
pretrain_monthly_error3 = pd.read_csv("../river-dl/pretrain_3/seasonal_month_metrics.csv")
pretrain_monthly_error4 = pd.read_csv("../river-dl/pretrain_4/seasonal_month_metrics.csv")
pretrain_monthly_error5 = pd.read_csv("../river-dl/pretrain_5/seasonal_month_metrics.csv")

PBinputs_monthly_error1 = pd.read_csv("../river-dl/no_pretrain_1/seasonal_month_metrics.csv")
PBinputs_monthly_error2 = pd.read_csv("../river-dl/no_pretrain_2/seasonal_month_metrics.csv")
PBinputs_monthly_error3 = pd.read_csv("../river-dl/no_pretrain_3/seasonal_month_metrics.csv")
PBinputs_monthly_error4 = pd.read_csv("../river-dl/no_pretrain_4/seasonal_month_metrics.csv")
PBinputs_monthly_error5 = pd.read_csv("../river-dl/no_pretrain_5/seasonal_month_metrics.csv")

PBinputs_monthly_error1_300 = pd.read_csv("../river-dl/no_pretrain_1_300/seasonal_month_metrics.csv")
```

```python
def plot_metric_matrix(tables, metric):
    matrix = np.zeros([len(tables), 12]) # 12 months
    count = 0
    for table in tables:
        table_metrics = table[metric]
        matrix[count, :] = table_metrics
        count += 1
    return(matrix)
```

```python
metric_matrix = plot_metric_matrix([pretrain_monthly_error1,
                                    pretrain_monthly_error2,
                                    pretrain_monthly_error3,
                                    pretrain_monthly_error4,
                                    pretrain_monthly_error5,
                                    PBinputs_monthly_error1,
                                    PBinputs_monthly_error2,
                                    PBinputs_monthly_error3,
                                    PBinputs_monthly_error4,
                                    PBinputs_monthly_error5,
                                    PBinputs_monthly_error1_300],
                                   'rmse')
plt.figure(figsize = (11.5, 11.5))
plt.imshow(metric_matrix)
plt.yticks([])
plt.xticks(np.linspace(-0.5, 11.5, 13),
           ["Dec\n21", "Jan\n21", "Feb\n21", "Mar\n21", "Apr\n21", "May\n21", "June\n21", "July\n21", "Aug\n21", "Sept\n21", "Oct\n21", "Nov\n21",
            "Dec\n21"]);

for i in range(metric_matrix.shape[0]):
    for j in range(metric_matrix.shape[1]):
        if metric_matrix[i, j] >= 10:
            plt.text(j-0.35, i+0.1, np.round(metric_matrix[i, j], 2), fontsize = 12)
        elif metric_matrix[i, j] >= 7:
            plt.text(j-0.25, i+0.1, np.round(metric_matrix[i, j], 2), fontsize = 12)
        else:
            plt.text(j-0.25, i+0.1, np.round(metric_matrix[i, j], 2), color = 'white', fontsize = 12)        
plt.text(-1, 2.75, 'PB pretraining', rotation = 90, fontsize = 14)
plt.text(-1, 8.125, 'PB inputs', rotation = 90, fontsize = 14)
plt.plot([-1.5, -0.5], [4.5, 4.5], color = 'black')
plt.plot([-0.5, 11.5], [4.5, 4.5], color = 'white')
plt.xlim(-1.5, 11.5)


#cbar = plt.colorbar(shrink = 0.7)

plt.title('Validation Set RMSE by Month \n Extrapolation Experiment: Training without Summer Data', fontsize = 18, y = 1.01)
plt.savefig("RMSE_by_month_PB_inputs_vs_pretraining.png", dpi = 300);
```
